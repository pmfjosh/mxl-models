```python
from mxlmodels import Simulator, data, get_pfennig2024_synechocystis, plot

dfs = data.pfennig2024.load()

model = get_pfennig2024_synechocystis(
    light_spectrum=dfs.light_spectrum,
    light_spectrum_measure=dfs.light_spectrum_measure,
    ocp_absorption=dfs.ocp_absorption_per_wavelength,
    abs_coef=dfs.pigment_abs_coef_per_wavelength,
    molar_masses=dfs.molar_masses,
    ps_comp=dfs.ps_comp,
    pigment_content=dfs.pigment_content,
)

res = Simulator(model).simulate(10).get_result().unwrap_or_err()
fig, ax = plot.lines(res.fluxes.loc[:, ["vPS2"]])
```
