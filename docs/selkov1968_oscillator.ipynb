```python
from mxlmodels import Simulator, get_selkov1968_glycolysis_oscillator, plot

res = (
    Simulator(get_selkov1968_glycolysis_oscillator())
    .simulate(10)
    .get_result()
    .unwrap_or_err()
)

fig, (ax1, ax2) = plot.two_axes(figsize=(12, 6))
_ = plot.lines(res.variables.iloc[:, :10], ax=ax1)
_ = plot.lines(res.fluxes.iloc[:, :10], ax=ax2)
```
